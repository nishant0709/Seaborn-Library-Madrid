```python
import seaborn as sns
```

```python
#JointPlot- A joint plot allow to study the relationship between 2 numeric variable. the central chart display their correlattion .it is usually a scatterplot a hexibin plot , a 2D histogram or a 2D desity plot
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#Pair plot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
#Distribution plot
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#Categorical plot
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#Box Plot
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='husl')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
plt.show()
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
