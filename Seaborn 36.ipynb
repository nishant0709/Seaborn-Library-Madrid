```python
import seaborn as sns
import pandas as pd
```

```python
sns.set(style='dark')
tips= sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip', data=tips,height=4,aspect=3)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip', hue="sex",data=tips,height=4,aspect=3)
```

```python
tips["size"].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip', hue="size",data=tips,height=4,aspect=3)
```

```python
sns.relplot(x='total_bill',y='tip',style='size', hue="sex",data=tips,height=4,aspect=3)
```

```python
sns.relplot(x='total_bill',y='tip', hue="smoker",style='time',size='size',data=tips,height=4,aspect=3)
```

```python
sns.relplot(x='total_bill',y='tip', hue="smoker",col='sex',data=tips,height=5,aspect=1)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='smoker',size='size',style='time')
```
